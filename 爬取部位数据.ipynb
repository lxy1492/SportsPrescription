```python
targets = ["胸部", "背部", "肩部", "手臂", "颈部", "腹部", "腰部", "臀部", "腿部", "全身"]
```

```python
target_URL = "https://www.gotokeep.com/trainingpoint/54826e417fb786000069ad82?keyword="
```

```python
def get_target_URL(key):
    return target_URL + key
```

```python
def getHTML(url:str):
    re = requests.get(url)
    content = re.content
    html = content.decode("utf-8")
    return html
```
