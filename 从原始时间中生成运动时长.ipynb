```python
from Moudle.sportMoudle import Sport
from Prescription.manage import getSport
```

```python
sports = getSport.loadAllSports()
```

```python
for each in sports:
    if isinstance(each,Sport):
        print(each.sportsTime)
```

```python
for each in sports:
    if isinstance(each,Sport):
        if '"' in each.sportsTime:
            t = each.sportsTime.replace('"',"")
            t1 = t.split("x")[0]
            t2 = t.split("x")[1]
            t1 = int(t1)
            t2 = int(t2)
            each.sportsTime = t1*t2
        else:
            t1 = each.sportsTime.split("x")[0]
            t2 = each.sportsTime.split("x")[1]
            t1 = int(t1)
            t2 = int(t2)
            each.sportsTime = t1*t2*15
```

```python
for each in sports:
    if isinstance(each,Sport):
        each.sportsTime = each.sportsTime/3
```
