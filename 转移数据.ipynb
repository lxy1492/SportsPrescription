```python
import os, json
```

```python
DataDIR = "./DataBase/Sports/Dict/"
```

```python
files = []
for each in os.listdir(DataDIR):
    if each[-5:] == ".json":
        files.append(each)
```

```python
datas = []
for each in files:
    f = open(os.path.join(DataDIR, each).replace("\\", "/"), "r")
    datas.append(json.loads(f.read()))
    f.close()
```

```python
for i, s in enumerate(datas):
    print(i, ">>>", s["name"])
```
