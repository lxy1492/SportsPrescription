```python
import os,pickle
from Moudle.courseMoudle import Course
from Moudle.sportMoudle import Sport
```

```python
courseDIR = "./DataBase/Course/"
sportDIR = "./DataBase/Sports/"
```

```python
courseList = []
for each in os.listdir(courseDIR):
    if each.split(".")[-1] == "tbs":
        courseList.append([each,courseDIR+each])
sportList = []
for each in os.listdir(sportDIR):
    if each.split(".")[-1] == "tbs":
        sportList.append([each,sportDIR+each])
```

```python
courses = []
sports = []
for each in courseList:
    with open(each[1],"rb") as f:
        courses.append(pickle.loads(f.read()))
for each in sportList:
    with open(each[1],"rb") as f:
        sports.append(pickle.loads(f.read()))
```

```python
for each in sports:
    print(each)
    print(each.detail)
```

```python
point = ["要点","呼吸","步骤","常见错误","细节图示"]

def modifyDetail(detail):
    d = {}
    for i in range(len(detail)):
        if detail[i] in point:
            j = i+1
            t = ""
            while(True):
                if j>=len(detail)-1:
                    break
                t += detail[j]
                j+=1
                if detail[j] in point:
                    break
            d.update({detail[i]:t})
    return d
```

```python
test = ['高提膝', '步骤', '自然站立，一侧腿微微踮起脚后跟，另一侧腿提膝去触碰对侧手心', '保持躯干稳定，双腿交替匀速进行', '呼吸', '自然呼吸', '动作感觉', '整个动作协调流畅', '常见错误', '错误：身体向前趴', '解决：稳定躯干，放慢速度', '细节图示']
```

```python
print(modifyDetail(test))
```

```python
for each in sports:
    d = modifyDetail(each.detail)
    d["细节图示"] = each.image
    each.detail = d
    each.save()
```
