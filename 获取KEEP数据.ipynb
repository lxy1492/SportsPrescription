```python
import os,json,requests,pickle
from lxml import etree
from multiprocessing import Process
from threading import Thread
from Moudle.courseMoudle import Course
from Moudle.sportMoudle import Sport

URL = "https://www.gotokeep.com/training"


def getHtml(url):
    res = requests.get(url)
    try:
        html = res.content.decode()
        return html
    except:
        try:
            html = res.content.decode("gb2312")
            return html
        except:
            return None
    return None
```

```python
def findSession(url=URL):
    res  =requests.get(url)
    try:
        html = res.content.decode()
    except:
        try:
            html = res.content.decode("gb2312")
        except:
            return -1
    element1 = etree.HTML(html)
    xpath1 = '//div[@class="keep-main-content"]//div[@class="keep-wrapper training"]//ul[@class="workout-hashtag clearfix"]//li//a[@class="keep-lazy"]//div[@class="title"]//text()'
    xpath2 = '//div[@class="keep-main-content"]//div[@class="keep-wrapper training"]//ul[@class="workout-hashtag clearfix"]//li//a[@class="keep-lazy"]//@href'
    titles = element1.xpath(xpath1)
    href = element1.xpath(xpath2)
    # print(len(titles),len(href))
    nextHtmlUrl = []
    for i in range(len(titles)):
        nextHtmlUrl.append([titles[i],"https://gotokeep.com"+href[i]])
    return nextHtmlUrl
```

```python
def getSubSession(url="https://gotokeep.com/workouthashtags/5e54e6450e9c352c75052408"):
    nameXpath = '//div[@class="keep-main-content"]//div[@class="keep-wrapper"]//ul[@class="courses clearfix"]//li//a//div[@class="name"]//text()'
    countXpath = '//div[@class="keep-main-content"]//div[@class="keep-wrapper"]//ul[@class="courses clearfix"]//li//a//div[@class="count"]//text()'
    levelXpath = '//div[@class="keep-main-content"]//div[@class="keep-wrapper"]//ul[@class="courses clearfix"]//li//a//div[@class="level"]//text()'
    poinnerXpath = '//div[@class="keep-main-content"]//div[@class="keep-wrapper"]//ul[@class="courses clearfix"]//li//a//div[@class="pioneer"]//text()'
    hrefXpath = '//div[@class="keep-main-content"]//div[@class="keep-wrapper"]//ul[@class="courses clearfix"]//li//a//@href'
    # res = requests.get(url)
    html = getHtml(url)
    if html==None:
        return html
    element = etree.HTML(html)
    name = element.xpath(nameXpath)
    count = element.xpath(countXpath)
    level = element.xpath(levelXpath)
    poinner = element.xpath(poinnerXpath)
    href = element.xpath(hrefXpath)
    counts = []
    i = 0
#     print(count)
    while(i<=len(count)-1):
#         print(i)
        e = i+1
        if e>=len(count):
            e = len(count)-1
        counts.append([count[i],count[e]])
        i+=2
    d = {}
    for i in range(len(name)):
        d.update({name[i]:{
            "name":name[i],
            "count":counts[i],
            "level":level[i],
            "poineer":poinner[i],
            "href":"https://gotokeep.com"+href[i],
        }})
    return d
```

```python
course= {}
for each in originalList:
    c = getSubSession(each[1])
    course.update({
        each[0]:c
    })
```

```python
def getSports(url="https://gotokeep.com/plans/5603ebfa331e338262f3d35d"):
    html = getHtml(url)
    if html==None:
        return html
    nameXpath = '/html/body/div[2]/div[1]/div[1]/div[1]/div//text()'
    descriptionXpath = '/html/body/div[2]/div[1]/div[1]/div[1]/p[1]//text()'
    countXpath = '/html/body/div[2]/div[1]/div[1]/div[1]/p[2]//text()'
    difficultyXpath = '/html/body/div[2]/div[1]/div[1]/div[1]/p[3]//text()'
    pioneerXpath = '/html/body/div[2]/div[1]/div[1]/div[1]/p[4]//text()'
    sportsnameXpath = '/html/body/div[2]/div[1]/div[2]/div[2]/div//p[@class="name"]//text()'
    sportstimeXpath = '/html/body/div[2]/div[1]/div[2]/div[2]/div//p[@class="payload"]//text()'
    hrefXpath = '/html/body/div[2]/div[1]/div[2]/div[2]/div//a//@href'
    backgroundimageXpath = '/html/body/div[2]/div[1]/div[2]/div[2]/div//a//div//@data-background'
    element = etree.HTML(html)
    name = element.xpath(nameXpath)
    description = element.xpath(descriptionXpath)
    count = element.xpath(countXpath)
    difficulty = element.xpath(difficultyXpath)
    pioneer = element.xpath(pioneerXpath)
    sportsname = element.xpath(sportsnameXpath)
    sportstime = element.xpath(sportstimeXpath)
    href = element.xpath(hrefXpath)
    backimage = element.xpath(backgroundimageXpath)
    sports = {}
    for i in range(len(sportsname)):
        # print(sportsname[i])
        sports.update({
            sportsname[i]:
            {
                "name": sportsname[i],
                "time": sportstime[i],
                "href": "https://gotokeep.com" + href[i],
                "image": "https://gotokeep.com" + backimage[i],
            }
        })
    return {
        "name":name,
        "description":description,
        "count":count,
        "difficulty":difficulty,
        "pioneer":pioneer,
        "sports":sports,
    }
```

```python
for each in data:
#     print(data[each])
    if len(data[each]["name"])<=0:
        break
    courseData = Course(name=data[each]["name"][0])
    courseData.setDescription(data[each]['description'][0])
    courseData.setDifficulty(data[each]['difficulty'][0].replace("K",""))
    courseData.setPioneer(data[each]["pioneer"][0])
    courseData.setSaveDIR("./DataBase/Course/")
#     print(courseData.saveDIR)
    for eachSport in data[each]["sports"]:
        filePath = "./DataBase/Sports/"+eachSport+".tbs"
        if os.path.exists(filePath):
            with open(filePath,"rb") as f:
                sp = pickle.loads(f.read())
        else:
            sp = Sport(name=eachSport,sportsTime=data[each]["sports"][eachSport]["time"],image=data[each]["sports"][eachSport]["href"])
        sp.setBelong(each)
        sp.setSaveDIR("./DataBase/Sports/")
#         print(sp.saveDIR)
        detailUrl = sp.image
        if not os.path.exists("./DataBase/Sports/"+eachSport+".tbs"):
            
            detail,imageSrc = getDetail(detailUrl)

            sp.detail = detail
            try:
                imageSrc = imageSrc[0]
                r = requests.get(imageSrc)
                if not os.path.exists("./DataBase/sportsImage/"):
                        os.makedirs("./DataBase/sportsImage/")
                with open("./DataBase/sportsImage/"+eachSport+".jpg","wb") as f:
                    f.write(r.content)
                sp.image = "./DataBase/sportsImage/"+eachSport+".jpg"
            except:
                sp.image = imageSrc
                
        sp.save()
        courseData.setSports(sp)
    courseData.save()
```

```python
def getDetail(url):
    # 详情介绍
    xpath1 = '//div[@class="keep-main-content"]//div[@class="exercise"]//div[@class="content"]//text()'
    # 示意图
    xpath2 = '//div[@class="keep-main-content"]//div[@class="exercise"]//div[@class="content"]//div[@class="img"]//img/@src'
    html = getHtml(url)
    element = etree.HTML(html)
    detail = element.xpath(xpath1)
    image_src = element.xpath(xpath2)
    return detail,image_src
```
